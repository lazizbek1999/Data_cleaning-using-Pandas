```python
cafe.columns = cafe.columns.str.strip().str.capitalize().str.replace(r' ','_')
```

```python
cafe['Item']=cafe['Item'].replace(['nan','UNKNOWN','ERROR'],None)
```

```python
cafe['Quantity']=pd.to_numeric(cafe['Quantity'],errors='coerce')
cafe['Price_per_unit']=pd.to_numeric(cafe['Price_per_unit'],errors='coerce')
cafe['Total_spent']=pd.to_numeric(cafe['Total_spent'],errors='coerce')
```

```python
cafe['Transaction_date']=pd.to_datetime(cafe['Transaction_date'],errors='coerce')
```

```python
cafe.info()
```

```python
cafe['help_col']=cafe['Price_per_unit']*cafe['Quantity']
mapping= cafe.dropna(subset='Item').set_index('help_col')['Item'].to_dict()
cafe['Item'] = cafe.apply(
lambda row:
    mapping.get(row['help_col'],row['Item']),axis=1
)
cafe.drop(columns=['help_col'],inplace=True)
```

```python
cafe['Quantity']=cafe.apply(
lambda r: r['Total_spent'] / r['Price_per_unit'] if pd.isna(r['Quantity']) else r['Quantity'],axis=1
)
```

```python
cafe['Price_per_unit']=cafe.apply(
lambda r: r['Total_spent'] / r['Quantity'] if pd.isna(r['Price_per_unit']) else r['Price_per_unit'],axis=1
)
```

```python
cafe['Total_spent']=cafe.apply(
lambda r: r['Price_per_unit'] * r['Quantity'] if pd.isna(r['Total_spent']) else r['Total_spent'],axis=1
)
```

```python
cafe[['Quantity','Price_per_unit','Total_spent']].isna().value_counts()
```

```python
cafe[['Quantity','Price_per_unit','Total_spent']]=cafe[['Quantity','Price_per_unit','Total_spent']].fillna(0)
```

```python
cafe['Payment_method']=cafe['Payment_method'].replace(['UNKNOWN','ERROR'],None)
```

```python
cafe['Location']=cafe['Location'].replace(['UNKNOWN','ERROR'],None)
```

```python
cafe['new_col']=cafe['Price_per_unit']*cafe['Quantity']
mapping= cafe.dropna(subset='Payment_method').set_index('new_col')['Payment_method'].to_dict()
cafe['Payment_method'] = cafe.apply(
lambda row:
    mapping.get(row['new_col'],row['Payment_method']),axis=1
)
cafe.drop(columns=['new_col'],inplace=True)
```

```python
cafe['new_col2']=cafe['Price_per_unit']*cafe['Quantity']
mapping= cafe.dropna(subset='Location').set_index('new_col2')['Location'].to_dict()
cafe['Location'] = cafe.apply(
lambda row:
    mapping.get(row['new_col2'],row['Location']),axis=1
)
cafe.drop(columns=['new_col2'],inplace=True)
```

```python
cafe['Transaction_date']=np.where(
    cafe['Transaction_date']=='nothing',
    None,
    cafe['Transaction_date']
)
```
